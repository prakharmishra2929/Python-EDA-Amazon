```python
# Checking B2B Data  by using pie chart 
B2B_check = df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_check, labels = B2B_check, autopct='%1.1f%%')
plt.show()
```

```python
# Checking B2B Data  by using pie chart 
B2B_check = df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_check, labels = B2B_check.index, autopct='%1.1f%%')
plt.show()
```

```python
#Note : From above chart we can see that maximum i.e. 99.3% of buyers are retailers and 0.7% are B2B buyers
```

```python
#  Prepare fullfilment data for pie chart
a1 = df['Fulfilment'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
#Scatter plot 
x = df['Category']
y = df['Size

plt.scatter(x,y)
plt.xlabel('Categories')
plt.ylabel("Size")
plt.title("Scatter Plot")
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize = (12,6))
sns.countplot(data = df,x="ship-state",color = 'red')
plt.title("Distribution")
plt.xlabel("state")
plt.ylabel("Count")
plt.xticks(rotation = 90)
plt.show()
```

```python
#top 10 states
top_10_states = df['ship-state'].value_counts().head(10)
plt.figure(figsize = (12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state',hue = 'ship-state')
plt.title("Distribution")
plt.xlabel('State')
plt.ylabel('count')
plt.xticks(rotation = 45)
plt.show()
```

```python
#Note: From above Graph you can see that most of the buyers are Maharashtra state
```
